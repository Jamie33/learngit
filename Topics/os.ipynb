### 当前路径及路径下的文件

os.getcwd()：查看当前所在路径。get current working director

os.listdir(path):列举目录下的所有文件。返回的是列表类型。list director

### 给文件重命名

os.rename(old_path, new_path)

os.rename(old_file_path, new_file_path), 只能对相应的文件进行重命名, 不能重命名文件的上级目录名.

os.renames(old_file_path, new_file_path), 是os.rename的升级版, 既可以重命名文件, 也可以重命名文件的上级目录名.

### 路径符号

**特别要注意：**

　　- win：路径全部都是反斜杠的，如：C:\Users\win7\learngit\Life

　　- linux:路径全部都是斜杠的，如：C:/Users/win7/learngit/Life

　　- Python中，所有路径都要采用Linux系统下的路径格式，用斜杠/
  

os.sep：输出操作系统特定的路径分隔符，win下为"\\", linux 下为 "/"

os.linesep：输出当前平台使用的行终止符，win下为 "\r\n"，Linux下为"\n"

os.pathsep：输出用于分割文件路径的字符串 ，win下为；  ，Linux 下为 ：

os.name：输出字符串指示当前使用平台。 win下为 "nt" ； Linux下为"posix"

os.system("bash command") ： 运行shell命令，直接显示

os.environ ：获取系统环境变量


str()： 函数返回一个用户易读的表达形式。

repr()： 产生一个解释器易读的表达形式。

### 判断一个文件是目录还是文件

```

# 判断文件是目录
Path('file_path').is_dir()

# 判断文件是个文件
Path('file_path').is_file()

```

````python
# 删除文件名带日期‘2017-01  2018.01 ’的字样

import re

path = 'F:\下载书籍'
pattern = r'\d{4}[-.]\d{1,2}'


for name in os.listdir(path):
    if re.search(pattern,name):
        oldpath = os.path.join(path,name)
        # print(re.search(pattern,name).group())
        print(name)
        newname = re.sub(pattern,'',name)
        print(newname)
        newpath = os.path.join(path,newname)
        os.rename(oldpath,newpath)
    else:
        print('无搜索结果')
````

````python
# 删除文件名带‘微信jnztxy’的字样
import re

pattern = '微信jnztxy'
path = 'F:\下载书籍'
print('当前目录 '+os.getcwd())

for name in os.listdir(path):
    if re.search(pattern,name):
        oldpath = os.path.join(path,name)
        #print(re.search(pattern,name).group())
        newname = re.sub(pattern,'',name)
        print(newname)
        newpath = os.path.join(path,newname)
        os.rename(oldpath,newpath)
        print(name)
````

````python
# 替换文件名 把大写数字替换成小写数字
import re

def transmes(path,intab,outtab):
  
    for name in os.listdir(path):
   
        oldpath = os.path.join(path,name)
        trantab = str.maketrans(intab,outtab)
        newname = name.translate(trantab) 
        newname = re.sub('零|十','',newname)
        newpath = os.path.join(path,newname)
        os.rename(oldpath,newpath)


intab = '百一二三四五六七八九'
outtab ='0123456789'
path = '/Users/xusanshan/learngit/Life/Comics/妖神记'

transmes(path,intab,outtab)
````
