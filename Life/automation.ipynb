## appium 爬取微信朋友圈

```python
from appium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from pymongo import MongoClient
from appium.webdriver.common.touch_action import TouchAction

PLATFORM = 'Android'
DEVICE_NAME = 'MI_PAD_4'
APP_PACKAGE = 'com.tencent.mm'
APP_ACTIVITY = 'ui.LauncherUI'
DRIVER_SERVER = 'http://localhost:4723/wd/hub'
TIMEOUT = 600
NORESET = 'True'
MONGO_URL = 'localhost'
MONGO_DB = 'moments'
MONGO_COLLECTION = 'moments'
FLICK_STRAT_X = 300
FLICK_START_Y = 300
FLICK_DISTANCE = 700

class Moments():
    
    def __init__(self):
        self.desired_caps = {
            'platformName':PLATFORM,
            'deviceName': DEVICE_NAME,
            'appPackage': APP_PACKAGE,
            'appActivity': APP_ACTIVITY,
            'noReset': NORESET,
            'newCommandTimeout': TIMEOUT,
        }
        
        print('打开 appium 服务器...')
        print('配置 appium ...')
        self.driver = webdriver.Remote(DRIVER_SERVER,self.desired_caps)
        self.wait = WebDriverWait(self.driver,TIMEOUT)
        print('打开 App ...等待{}s...'.format(TIMEOUT))
        self.client = MongoClient(MONGO_URL)
        self.db = self.client[MONGO_DB]
        self.collection = self.db[MONGO_COLLECTION]

    def enter(self):
        print('点击发现...')
        tab = self.wait.until(EC.presence_of_element_located((By.XPATH,'//*[@resource-id="com.tencent.mm:id/bq"]/android.widget.LinearLayout/android.widget.RelativeLayout[3]/android.widget.LinearLayout')))
        tab.click()
        print('进入朋友圈...')
        moments = self.wait.until(EC.presence_of_element_located((By.XPATH,'//*[@resource-id="android:id/list"]/android.widget.LinearLayout[1]')))
        moments.click()
        
    def crawl(self):
        while True:
            self.driver.swipe(FLICK_STRAT_X,FLICK_START_Y+FLICK_DISTANCE,FLICK_STRAT_X,FLICK_START_Y)
            items = self.wait.until(EC.presence_of_all_elements_located((By.ID,'com.tencent.mm:id/ej_')))
            for item in items:
                try:
                    nickname = item.find_element_by_id('com.tencent.mm:id/b5o').get_attribute('text')
                    content = item.find_element_by_id('com.tencent.mm:id/ejc').get_attribute('text')
                    date = item.find_element_by_id('com.tencent.mm:id/eec').get_attribute('text')
                    print(nickname,content,date)
                except NoSuchElementException:
                    pass
            
    def main(self):
        self.enter()     
        self.crawl()
        
if __name__ == '__main__':
    moments = Moments()
    moments.main()
```
