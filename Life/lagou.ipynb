```python
from selenium import webdriver
import time,re,csv,pymongo,os
from pyquery import PyQuery as pq
import requests
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
import selenium.webdriver.support.expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains 
from selenium.webdriver.common.keys import Keys


def get_job(driver,job_url):
    driver.get(job_url) 
    time.sleep(2)
    for i in range(2):
        driver.find_element_by_xpath("/html/body").send_keys(Keys.PAGE_DOWN)
        time.sleep(1)
    html = driver.page_source
    doc = pq(html,parser="html")
    job_detail = doc('.job-detail').text()
    if '英语' in job_detail:
        job_name = doc('.job-name').text()
        print('job url: {}\njob name: {}\njob detail:\n{}'.format(job_url,job_name,job_detail))
    elif '翻译' in job_detail:
        job_name = doc('.job-name').text()
        print('job url: {}\njob name: {}\njob detail:\n{}'.format(job_url,job_name,job_detail))           
    
def get_list(driver,list_url):
    driver.get(list_url) 
    time.sleep(2)
    for i in range(6):
        driver.find_element_by_xpath("/html/body").send_keys(Keys.PAGE_DOWN)
        time.sleep(1)
    html = driver.page_source
    doc = pq(html,parser="html")
    job_list = doc('.con_list_item.default_list')
    for job in job_list.items():
        job_num = job.attr('data-positionid')
        job_url = 'https://www.lagou.com/jobs/{}.html'.format(job_num)
        get_job(driver,job_url)

def get_page(page,driver,url):
    for p in range(1,page+1):
        print('current page {}'.format(p))
        if p == 1:
            list_url = home_url
        else: 
            list_url = home_url + str(p)
        get_list(driver,list_url)
        
        
#job_url = r'https://www.lagou.com/jobs/5449832.html'               
home_url = 'https://www.lagou.com/zhaopin/Python/'                    
page = 30

if __name__ == '__main__':
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--headless")
    driver = webdriver.Chrome(chrome_options=chrome_options)
    driver.maximize_window()    
    get_page(page,driver,home_url)
    driver.close()
```
