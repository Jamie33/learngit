## appium 趣头条 

````python
from appium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time,random

class qudoutiao():
    
    def __init__(self):
        self.desired_caps = {
            'platformName':platformName,
            'deviceName':deviceName,
            'appPackage':appPackage,
            'appActivity':appActivity,
            'noReset': "True",
            'newCommandTimeout': 600,
        }
        
        print('打开 appium 服务器...')
        print('配置 appium...')
        self.driver = webdriver.Remote(server,self.desired_caps)
        self.wait = WebDriverWait(self.driver,timeout)
        print('打开趣头条...等待{}s...'.format(timeout))
  
    def login(self):
        print('点击弹窗...先去逛逛...')
        popup = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/aju')))
        popup.click()
        print('进入底部菜单:我的...')        
        login = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/la')))
        login.click()
        print('点击手机登录按钮...')
        mobile_login = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/az_')))
        mobile_login.click()
        print('隐藏键盘...')
        self.driver.hide_keyboard()
        print('点击密码登录按钮...')
        account_login = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/aza')))
        account_login.click()
        print('输入手机号码...')
        enter_mobile = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/asm')))
        enter_mobile.send_keys(mobile)
        print('输入密码...')
        password_button = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/aso')))
        password_button.send_keys(password)
        print('隐藏键盘...')
        self.driver.hide_keyboard()       
        print('点击登录按钮...')
        submit = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/o8')))
        submit.click()
        print('点击体验弹窗...')
        exp = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/a6h')))
        exp.click()
        
 

    def view(self):
        #print('点击弹窗...先去逛逛...')
        #popup = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/aju')))
        #popup.click()        
        print('点击底部菜单:视频...')
        video = self.wait.until(EC.presence_of_element_located((By.ID, 'com.jifen.qukan:id/ks')))
        video.click()
        t = 1
        while True:
            content = self.driver.find_elements_by_id("com.jifen.qukan:id/a4d")
            time_text = self.driver.find_elements_by_id("com.jifen.qukan:id/a6e")
            print('第{}次进入循环...'.format(t))
            for i in range(len(content)):
                try:
                    time_span = time_text[i].get_attribute('text').split(':')
                    content[i].click()
                    print('点击第{}视频'.format(i+1))
                    time_cost = int(time_span[0])*60+int(time_span[1])
                    time.sleep(random.randint(4,10))
                    print('该视频时长:{}s'.format(time_cost))
                    time.sleep(time_cost) 
                except IndexError:
                    print('此为广告...')
                    continue
                    
            print('点击刷新页面...')
            self.driver.find_element_by_id("com.jifen.qukan:id/ks").click()
            t += 1

platformName = 'Android'
deviceName = '127.0.0.1:5555'
appPackage = 'com.jifen.qukan'
appActivity = 'com.jifen.qkbase.main.MainActivity'
server = 'http://localhost:4723/wd/hub'
timeout = 60
mobile = ""
password = ""

q = qudoutiao()
q.view()
````
