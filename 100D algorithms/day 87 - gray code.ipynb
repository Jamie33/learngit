```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
```

## algorithm

```python
def gray_code(n):
    code = [0] * n
    i, parity = 0, 0

    while i < n:
        yield code[::-1]

        i = parity
        parity ^= 1
        while i and not code[i - 1]:
            i += 1

        if i < n:
            code[i] ^= 1            
```

## run

```python
for code in gray_code(4):
    print(code)
```

```python
def plot_code(n):
    code = np.array(list(gray_code(n))).T

    plot = figure(x_range=(0, 1), y_range=(0, 1), plot_width=800, plot_height=200)
    plot.axis.visible = False
    plot.image([code], x=0, y=0, dw=1, dh=1, palette=['#ffffff', '#000000'])

    show(plot)
```

```python
output_notebook()
```

```python
plot_code(5)
```
