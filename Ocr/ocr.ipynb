## [百度翻译接口](http://api.fanyi.baidu.com/api/trans/product/index)

```python
#/usr/bin/env python
#coding=utf8
 
import requests,json
import urllib
import random
import hashlib

 
def translate(q,fromLang,toLang,appid,secretKey):
    myurl = 'http://api.fanyi.baidu.com/api/trans/vip/translate'
    salt = random.randint(32768, 65536)
    sign = appid+q+str(salt)+secretKey
    sign = hashlib.md5(sign.encode("utf-8")).hexdigest()
    myurl = myurl+'?appid='+appid+'&q='+urllib.request.quote(q)+'&from='+fromLang+'&to='+toLang+'&salt='+str(salt)+'&sign='+sign
    try:
        response = requests.get(myurl).content
        content = str(response,encoding = "utf-8")
        json_reads = json.loads(content)
        trans_text = ' \n '.join(i['dst'] for i in json_reads['trans_result'])
        return trans_text
    except Exception as e:
        print(e)

q = 'apple \n orange \n pineapple'
fromLang = 'en'
toLang = 'zh'
appid = '' #你的appid
secretKey = '' #你的密钥

translate(q,fromLang,toLang,appid,secretKey)
```
