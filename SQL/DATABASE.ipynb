<span class="girk">创建数据库</span>

CREATE DATABASE <数据库名称>;


CREATE DATABASE shop;


<span class="girk">创建表</span>


CREATE TABE <表名>

(

<列名1><数据类型><该列所需约束>,

<列名2><数据类型><该列所需约束>,

<列名3><数据类型><该列所需约束>,

<列名4><数据类型><该列所需约束>,
             .
             .
             .

<该表的约束1>,<该表的约束2>，...

);



CREATE TABLE Shopitem
(

item_id     CHAR(4)      NOT NULL,

item_name  VARCHAR(100)  NOT NULL,

item_catalog VARCHAR(32) NOT NULL,

item_price  INTEGER ,

item_cost INTEGER,

register_data DATA,

PRIMARY KEY (item_id)
);


<span class="girk">表的删除</span>

DROP TABLE <表名>;

DROP TABLE Shopitem;


<span class="girk">表定义的更新</span>


添加列的 ALTER TBABLE 语句

ALTER TABLE <表名> ADD COLUMN<列的定义>;

ALTER TABLE Shopitem ADD COLUMN item_quantity INTEGER;


删除列的 ALTER TBABLE 语句

ALTER TABLE <表名> DROP COLUMN<列名>;

ALTER TABLE Shopitem DROP COLUMN item_quantity;
