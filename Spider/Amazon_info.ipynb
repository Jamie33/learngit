## 产品信息

```python
from selenium import webdriver
import time,re,csv,random,datetime
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains
from pyquery import PyQuery as pq

def get_details(item_url,save_file):
    
    # Chrome object
    with open(save_file,'a',encoding='utf-8') as w:
        try:
            print('getting item details from Amazon ...')
 
            driver = webdriver.PhantomJS() 
            driver.get(item_url) 
            time.sleep(5)
            print('item url: {}'.format(item_url))
            content = driver.find_element_by_id('centerCol')
            item_name = content.find_element_by_id('productTitle').text
            item_name = re.sub(r',','',item_name)
            print('getting item name:{}'.format(item_name)) 
            try:
                review = content.find_element_by_id('acrCustomerReviewText').text
                review_db = re.search(r'(.+)customer review',review).group(1)
                review_num = re.sub(r',','',review_db)
                print('getting item reviews:{}'.format(review_num))    
            except NoSuchElementException :
                print('no item reviews')
                review_db = 0
                review_num = 0        
            price = content.find_element_by_id('priceblock_ourprice').text
            print('getting item price:{}'.format(price)) 
            w.write('{},{},{},{}\n'.format(item_name,price,review_num,item_url))
            driver.close()
            
        except TimeoutException as err:
            print(err)
            print('failed to load this url{},waitting for 10 seconds to reload'.format(item_url))
            time.sleep(10)
            get_details(item_url,save_file)
    print('getting item details done')
            

item_url = "https://www.amazon.com/Levis-Mens-Regular-Stonewash-34x32/dp/B0018OKNWM/ref=sr_1_1?s=apparel&ie=UTF8&qid=1548214872&sr=1-1&nodeID=1045564&psd=1"
save_file = "amazon_result.csv"


with open(save_file,'a',encoding='utf-8-sig', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Item name','Price','Review num','Item url'])


get_details(item_url,save_file)
```

## 获取列表url

```python
from selenium import webdriver
import time,re,csv,random,datetime
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains
from pyquery import PyQuery as pq
from selenium.webdriver.common.keys import Keys



def get_list(page_url,save_file):
    
    # Chrome object

    print('Getting item list from Amazon...')
    driver = webdriver.Chrome() 
    driver.maximize_window()
    driver.get(page_url) 
    time.sleep(5)
    
    url_list = [i.get_attribute('href') for i in driver.find_elements_by_css_selector('.a-link-normal.s-access-detail-page.s-overflow-ellipsis.s-color-twister-title-link.a-text-normal')]        
    print('item url list: {}'.format(len(url_list)))

    with open(save_file,'a',encoding='utf-8') as w:
        for i in url_list:
            w.write('{}\n'.format(i))
            
    print('Getting item list Done!')
    driver.close()
     
page_url = r'https://www.amazon.com/s/ref=sr_st_featured-rank?rh=n%3A7141123011%2Cn%3A7147441011%2Cn%3A1040658%2Cn%3A1045564&qid=1548214701&sort=featured-rank&ajr=3'
save_file = r'item_list.csv'

get_list(page_url,save_file)
```

### 获取前20页，所有商品

```python
from selenium import webdriver
import time,re,csv,random,datetime
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains
from pyquery import PyQuery as pq
from selenium.webdriver.common.keys import Keys



def get_list(driver,page_url,save_file,page_num):
    
    driver.get(page_url) 
    time.sleep(5)
    
    current_page = int(driver.find_element_by_css_selector('span.pagnCur').text)
    print('current page: No.{}, url: {}'.format(current_page,page_url))

    url_list = [i.get_attribute('href') for i in driver.find_elements_by_css_selector('.a-link-normal.s-access-detail-page.s-overflow-ellipsis.s-color-twister-title-link.a-text-normal')]        
    print('getting {} item urls'.format(len(url_list)))

    with open(save_file,'a',encoding='utf-8') as w:
        for i in url_list:
            w.write('{}\n'.format(i))
    
    if current_page < page_num:      
        next_page = driver.find_element_by_id('pagnNextLink').get_attribute('href') 
        print('navigate to next page: {} ...'.format(next_page))            
        get_list(driver,next_page,save_file,page_num)   
            

def get_page(page_url,save_file,page_num):
    
    print('Getting item urls from Amazon...')
    driver = webdriver.Chrome()
    time.sleep(random.randint(2,10))
    get_list(driver,page_url,save_file,page_num)
    print('Getting item list done!')
    driver.close()
    

page_url = r'https://www.amazon.com/s/ref=sr_st_featured-rank?rh=n%3A7141123011%2Cn%3A7147441011%2Cn%3A1040658%2Cn%3A1045564&qid=1548214701&sort=featured-rank&ajr=3'
save_file = r'20_page_item_list.csv'
page_num = 20

get_page(page_url,save_file,page_num)
```

## 读取csv中url (可行)

```python
from selenium import webdriver
import time,re,csv,random,datetime,pymongo
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains
from pyquery import PyQuery as pq

def get_details(driver,item_url,save_file):

    with open(save_file,'a',encoding='utf-8') as w:
        try:
            print('getting item details from Amazon ...')
             
            driver.get(item_url) 
            time.sleep(5)
            print('item url: {}'.format(item_url))
            #content = driver.find_element_by_id('centerCol')
            item_db = driver.find_element_by_id('productTitle').text
            item_name = re.sub(r',','',item_db)
            print('getting item name:{}'.format(item_name))
            
            try:
                review = driver.find_element_by_id('acrCustomerReviewText').text
                # review/reviews
                review_db = re.search(r'(.+)customer review',review).group(1)
                review_num = re.sub(r',','',review_db)
                print('getting item reviews:{}'.format(review_num))    
            except NoSuchElementException :
                print('no item reviews')
                review_db = 0
                review_num = 0
            
            try:
                price = driver.find_element_by_id('priceblock_ourprice').text
                price = re.sub(r',','',price)
                print('getting item price:{}'.format(price))  
            except NoSuchElementException:
                print('no price')
                price = 'no price'
                
            
            product = {
                'item_name': item_db,
                'price': price,
                'review_num': review_db,
                'item_url': item_url,
            }
            save_to_mongo(product)
            
            
            w.write('{},{},{},{}\n'.format(item_name,price,review_num,item_url))
            
        except (TimeoutException) as err:
            print(err)
            print('failed to load this url{},waitting for 10 seconds to reload'.format(item_url))
            time.sleep(10)
            get_details(driver,item_url,save_file)
    print('getting item details done')
            

def save_to_mongo(result):
    MONGO_URL = 'localhost'
    MONGO_DB = 'shopcrawl'
    MONGO_COLLECTION = 'amazon'
    client = pymongo.MongoClient(MONGO_URL)
    db = client[MONGO_DB]
    collection = db[MONGO_COLLECTION]
    try:
        if collection.insert_one(result):
            print('save to mongodb')
    except Exception:
        print('failed to save to mongodb')
        
        
def amazon(file_name,save_file):
    start = time.time()
    driver = webdriver.Chrome()
    

    with open(file_name,'r',encoding='utf-8') as f:
        i = 0
        reader = csv.reader(f)
        rows = [row for row in reader]
        for row in rows[766:]:
            time.sleep(random.randint(2,10))
            print('Getting item No.{}'.format(i))
            item_url = row[0]
            get_details(driver,item_url,save_file)
            i += 1
    driver.close()
    end = time.time()
    print('End in {} minutes'.format(i,(end-start)/60)) 
        
save_file = r'page20_item_details.csv'
file_name = r'page20_item_list.csv'

with open(save_file,'a',encoding='utf-8', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Item name','Price','Review num','Item url'])

amazon(file_name,save_file)
```
