## 从单个url 里面获取图片

```python
from selenium import webdriver
import time,re,os,requests
from PIL import Image
from io import BytesIO
from selenium.webdriver.chrome.options import Options

#url = 'https://www.lazada.co.id/products/xiaomi-original-headset-bluetooth-41-earphone-build-in-mic-handfree-i170124489-s199915361.html?spm=a2o4j.searchlist.list.7.673259f48YtVQX&search=1'


def get_list(url):
    print('Download start...')
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    driver = webdriver.Chrome(chrome_options=chrome_options)  #声明浏览器对象
    driver.get(url) # 访问网页
    print('Open page...')
    time.sleep(2)
    director = re.sub('[^a-z0-9A-Z_]','-',driver.title)[:20]
    print('Product name: {}'.format(director))
    pic_urls = driver.find_elements_by_css_selector('.item-gallery__image-wrapper img')
    num = 1
    print('Get pic list...')
    for i in pic_urls:
        pic_url = i.get_attribute('src')
        pic_url = re.sub('80x80','800x800',pic_url,count=0)
        save_image(pic_url,director,num)
        num += 1
        print('Save pic {}...'.format(num))
        

def save_image(img_url,director,num):
    
    img_response = requests.get(img_url)
    path = r'C:\Users\win7\learngit\Spider\Pictures\{}'.format(director)
    isExists = os.path.exists(path)
    if not isExists:
        os.makedirs(path)
        print('directory is already created')
    else:
        print('directory is already existed')
    with open('{}\{}.jpg'.format(path,num),'ab') as f:
        f.write(img_response.content)
    print('Pic {} complete'.format(num))
        
        
```

## 获取详情页图片

```python
from selenium import webdriver
import time,re,os,requests
from PIL import Image
from io import BytesIO
from selenium.webdriver.chrome.options import Options



url = 'https://www.lazada.co.id/products/karina-wardrobe-aluna-sling-bag-tas-wanita-tas-murah-tas-ransel-sling-bag-tas-bahu-wanita-tas-fashion-tas-cewek-tas-wanita-murah-tas-cantik-tas-branded-tas-kerja-tas-kuliah-tas-selempang-tas-slempang-wanita-slempang-i398395811-s449435701.html?spm=a2o4j.searchlistcategory.list.77.4bca5fd7E7tsPA&search=1'


    
def get_pics(url):
    
    
    # Chrome object

    print('Downloading pictures from Lazada ...')
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    driver = webdriver.Chrome(chrome_options=chrome_options)  
    driver.get(url) 
    print('Open page...')
    time.sleep(2)
    
    # Get prduct name, pictures
    
    director = re.sub('[^a-z0-9A-Z_]','-',driver.title)[:20]
    print('Handling product : {}'.format(director))
    pic_urls = driver.find_elements_by_css_selector('.item-gallery__image-wrapper img')
    description = driver.find_elements_by_css_selector('.pdp-product-desc ')
    
    
    # Product pictures
    
   
    for i in pic_urls:
        pic_url = i.get_attribute('src')
        pic_url = re.sub('80x80','800x800',pic_url,count=0)
        pro_name = os.path.basename(pic_url)
        save_image(pic_url,director,pro_name)
    

    # Pictures in description     
    
    for i in description:
        imgs = i.find_elements_by_tag_name('img')
        if len(imgs) != 0:
            for img in imgs:
                img_url = img.get_attribute('src')
                des_name = os.path.basename(img_url)
                save_image(img_url,director,des_name)
        else:
            print('This product has no pictures in description.')
    
    print('Downloading is compeleted.')
                


def save_image(img_url,director,name):
    print('Saving picture: {}'.format(name))
    img_response = requests.get(img_url)
    path = r'C:\Users\win7\learngit\Spider\Pictures\{}'.format(director)
    isExists = os.path.exists(path)
    if not isExists:
        os.makedirs(path)
    with open('{}\{}'.format(path,name),'ab') as f:
        f.write(img_response.content)


        
get_pics(url)
```
