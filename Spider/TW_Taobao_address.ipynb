## 淘宝台湾地址抓取 （可行）
字段：市、区

```python
from selenium import webdriver
import time
import datetime
from threading import Timer

class taobao():
    
    def __init__(self,target_url,save_file):
        self.item_url = target_url
        self.driver = webdriver.Chrome()
        self.save_file = save_file
    
    def login(self):
        url = 'https://login.taobao.com/member/login.jhtml'
        self.driver.get(url)
        self.driver.maximize_window()
        print("请在15秒内完成扫码登录")
        time.sleep(10)
        self.driver.get(target_url)
        print('进入地址')
        
    def switch(self):
        self.driver.refresh()
        print('切换')
        self.driver.find_element_by_xpath('//*[@id="cndzkEntrance"]/div[1]/div/div/a').click()
        print('台湾')
        self.driver.find_element_by_xpath('//*[@id="cndzkEntrance"]/div[1]/div/div[2]/div/ul[2]/li[4]').click()
        print('确定')
        self.driver.find_element_by_css_selector('.cndzk-entrance-country-box-button').click()
    
    def choose(self,num):
        print('选择地址')
        self.driver.find_element_by_css_selector('.cndzk-entrance-division-header-click').click()
        provinces = self.driver.find_elements_by_css_selector('.cndzk-entrance-division-box-content-tag')
        print('province number {}'.format(num))
        with open(self.save_file,'a',encoding='utf-8') as w:
            province = provinces[num].text
            provinces[num].click()
            areas = self.driver.find_elements_by_css_selector('.cndzk-entrance-division-box-content-tag')
            for a in areas:
                area = a.text
                print(province,area)
                w.write('{},{}\n'.format(province,area))
    
    def close(self):
        self.driver.close()
                

item_url = 'https://detail.tmall.com/item.htm?id=540107236593&spm=a21bz.7725273.1998564503.4.381c3db8HH19rn&umpChannel=qianggou&u_channel=qianggou'
save_file = r'Taiwan_address_test.csv'

if __name__ == '__main__':
    target_url = 'https://member1.taobao.com/member/fresh/deliver_address.htm?spm=a1z08.2.0.0.28da978bR4XHvj'
    t = taobao(target_url,save_file)
    t.login()
    for num in range(22):
        time.sleep(3)
        t.switch()
        t.choose(num)
    t.close()
    print('Done')
```
