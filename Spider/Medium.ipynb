```python
from selenium import webdriver
import time
import datetime
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.action_chains import ActionChains
from pyquery import PyQuery as pq


class medium():
    
    def __init__(self,target_url,save_file):
        self.item_url = target_url
        self.driver = webdriver.Chrome()
        self.driver.set_page_load_timeout(30)
        self.save_file = save_file
    
    def login(self):
        url = target_url
        self.driver.get(url)
        print("getting into link")

        
    def get_article(self):
        print('getting articles')
        html = self.driver.page_source      # 获取页面的源代码
        doc = pq(html)
        print(doc)
        class = 'r s y'
        

    def scroll_down(self):
        print('scrolling down to the bottom')
        self.driver.execute_script(""" 
            (function () { 
                var y = document.body.scrollTop; 
                var step = 100; 
                window.scroll(0, y); 
                function f() { 
                    if (y < document.body.scrollHeight) { 
                        y += step; 
                        window.scroll(0, y); 
                        setTimeout(f, 50); 
                    }
                    else { 
                        window.scroll(0, y); 
                        document.title += "scroll-done"; 
                    } 
                } 
                setTimeout(f, 1000); 
            })(); 
            """)
        time.sleep(50)
        print('waitting for 50 seconds')
        
    
    def close(self):
        self.driver.close()
                

item_url = ''
save_file = r'Taiwan_address_test.csv'

if __name__ == '__main__':
    target_url = 'https://medium.com/@luxchain'
    t = medium(target_url,save_file)
    t.login()
    t.scroll_down()
    t.get_article()
#     for num in range(22):
#         time.sleep(3)
#         t.switch()
#         t.choose(num)
#     t.close()
    print('Done')
```
