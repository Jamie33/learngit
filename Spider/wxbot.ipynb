```python
import requests
from pyquery import PyQuery as pq
import datetime


def get_price(keyword):
    if keyword.lower() == 'btc':
        keyword = 'bitcoin'
    elif keyword.lower() == 'eth':
        keyword = 'ethereum'
    elif keyword.lower() == 'xrp':    
        keyword = 'ripple'
    url = 'https://www.feixiaohao.com/currencies/{}/'.format(keyword.lower())
    doc = pq(url)
    nowTime =  datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    content = doc('.cell')
    info = content.find('.coinprice')
    fluctuation =info.find('.tags-red').text()
    info.find('.tags-red').remove()
    price = info.text()
    rank = content('.tag-marketcap').text()
    result = "非小号{}的价格:{}\n涨幅:{}\n市值排名:{}\n更多: http://fxh.io/bitcoin/\n【{}】".format(keyword,price,fluctuation,rank,nowTime)
    print(result)
    
get_price('btc')
```

```python
import requests
import datetime
from pyquery import PyQuery as pq


url = 'https://www.feixiaohao.com/'

coindict = {}

def get_name(url):     
    doc = pq(url)
    coinlist = doc('#table tbody')
    names = coinlist.children('tr')
   
    for tr in names.items():
        fullname = tr.attr('id')
        shortname = tr.children('td a img').attr('alt').split('-')[0]
        coindict[shortname]=fullname
    return coindict   

def get_page(url):
    doc = pq(url)
    pagelist = doc('.new-box-tit').children('.new-page-list').children('a')
    maxpage = list(pagelist.items())[-2].text()
    return maxpage



for i in range(24):
    url = 'https://www.feixiaohao.com/list_{}.html'.format(i+1)
    get_name(url)


#get_name('https://www.feixiaohao.com/currencies/list_1.html')
print(coindict)
```
