```python
from selenium import webdriver
import time,re,csv,random,datetime
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains

def get_details(item_url,save_file):
    
    # Chrome object
    with open(save_file,'a',encoding='utf-8') as w:
        try:
            print('getting item details from Shopee ...')
            chrome_options = Options()
            chrome_options.add_argument('--headless')
            driver = webdriver.Chrome(chrome_options=chrome_options)  
            driver.get(item_url) 
            wait = WebDriverWait(driver,10)
            
          
            wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.flex.items-center._1z1CEl.page-product__breadcrumb')))

            item_class = driver.find_elements_by_css_selector('div.flex.items-center._1z1CEl.page-product__breadcrumb a')
            print(item_url)
            platform = item_class[0].text
            print('getting item platform: {}'.format(platform))
            main_class = item_class[1].text
            print('getting item main class: {}'.format(main_class))
            sub_class = item_class[2].text
            print('getting item sub class: {}'.format(sub_class))
            pro_name = driver.find_element_by_css_selector('div.flex.items-center._1z1CEl.page-product__breadcrumb span').text
            item_name = re.sub('[^a-z0-9A-Z_]', '', pro_name, count=0)
            print('getting item name: {}'.format(item_name))

            price = driver.find_element_by_css_selector('div._3n5NQx').text
            print('getting current price: {}'.format(price))
        
            
            try:
                after_sale = driver.find_element_by_css_selector('a._13C8_x.flex.items-center').text
                print('after sale info: {}'.format(after_sale))
            except Exception as err:
                after_sale = 'N/A'
                print('did not get after sale info Error message:{}'.format(err))
            
            try:
                list_price = driver.find_element_by_css_selector('div._3_ISdg').text
                print('getting list price: {}'.format(list_price))
            except NoSuchElementException:
                list_price = price
                print('list price is equal to price')
            
            '''
           
            try:
                brand = driver.find_element_by_css_selector('._1ymsZN').text                
                print('getting brand: {}'.format(brand))
            except Exception as err:
                print('did not get brand Error message:{}'.format(err))
                brand = 'N/A'
            '''
            
            try:
                variations = driver.find_elements_by_css_selector('.product-variation')
                variation = '{};'*len(variations)
                for i in variations:
                    
                    print(i.text)
                print('getting variation')
            except Exception as err:
                print('did not get variation Error message:{}'.format(err))
                variation = 'N/A'
            
            
            try:
                description = driver.find_element_by_css_selector('div.product-detail.page-product__detail')
                print('getting description: {}'.format(description))
            except Exception as err:
                print('did not get description Error message:{}'.format(err))
                description = 'N/A'
                
                
            stock = 1000
            warning_code = 10
            shipping_fee = 15
      
            #print(platform,main_class,sub_class,item_name,list_price,price,after_sale,brand,description,item_url)
            #w.write('{},{},{},{},{},{},{},{},{},{},{},{}\n'.format(platform,main_class,sub_class,item_name,list_price,price,item_score,review_num,month_sales,favorite_num,shipping_note,item_url))
            driver.close()
            
        except TimeoutException as err:
            print(err)
            print('failed to load this url{},waitting for 10 seconds to reload'.format(item_url))
            time.sleep(10)
            get_details(item_url,save_file)
    print('getting item details done')
            

item_url = "https://shopee.co.id/Jam-Q-Q-Jelly-Angka-i.6684120.71543080"
save_file = "E:\Jamie\python test\csv\shopee_result.csv"
        
get_details(item_url,save_file)
```
