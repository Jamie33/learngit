```python
from selenium import webdriver
import time,re,csv,random,datetime,requests,os
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains
from PIL import Image
from selenium.webdriver.common.keys import Keys

headers = {'User-Agent': 'User-Agent:Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}

def get_shopee_details(item_url,save_file):
    
    with open(save_file,'a',encoding='utf-8') as w:
        try:
            print('getting item details from Shopee ...')
            driver = webdriver.PhantomJS()
            driver.maximize_window()
            driver.get(item_url)
            wait = WebDriverWait(driver,10)

            #jsCode = "var q=document.documentElement.scrollTop=100000"
            #driver.execute_script(jsCode)

            wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.flex.items-center._1z1CEl.page-product__breadcrumb')))
            item_class = driver.find_elements_by_css_selector('div.flex.items-center._1z1CEl.page-product__breadcrumb a')
            print('item url: {}'.format(item_url))
            platform = item_class[0].text
            print('getting item platform: {}'.format(platform))
            main_class = item_class[1].text
            print('getting item main class: {}'.format(main_class))
            sub_class = item_class[2].text
            print('getting item sub class: {}'.format(sub_class))
            
            category_info = ':'.join([i.text for i in item_class])
            print('getting item category：{}'.format(category_info))
            
            pro_name = driver.find_element_by_css_selector('div.flex.items-center._1z1CEl.page-product__breadcrumb span').text
            item_name = ' '.join(re.sub('[^a-z0-9A-Z_\-\s]','', pro_name, count=0).split())
            print('getting item name: {}'.format(item_name))
            
            
            try:
                aftersale_info = driver.find_element_by_css_selector('a._13C8_x.flex.items-center').text
                after_sale = re.sub(r'\n',' ',aftersale_info,count=0)
                print('getting after sale info: {}'.format(after_sale))
            except:
                try:
                    aftersale_info = driver.find_element_by_css_selector('div._3E_omo').text
                    after_sale = re.sub(r'\n',' ',aftersale_info,count=0)
                    print('getting after sale info: {}'.format(after_sale))
                except Exception as err:
                    after_sale = 'N/A'
                    print('did not get after sale info Error message:{}'.format(err))
            '''
            
            for i in range(5):
                time.sleep(2)
                driver.find_element_by_xpath("/html/body").send_keys(Keys.PAGE_DOWN)
            '''
    
            try:
                brand = driver.find_element_by_css_selector('a._2H-513').text
                print('getting brand: {}'.format(brand))
            except Exception as err:
                brand = 'N/A'
                print('did not get brand Error message:{}'.format(err))
            
            
            price = driver.find_element_by_css_selector('div._3n5NQx').text
            print('getting current price: {}'.format(price))

            try:
                list_price = driver.find_element_by_css_selector('div._3_ISdg').text
                print('getting list price: {}'.format(list_price))
            except NoSuchElementException:
                list_price = price
                print('list price is equal to price')

                
            print('getting pic urls')
            pic_urls = driver.find_elements_by_css_selector('._2Fw7Qu.animated-lazy-image__image--ready')
            n = 1
            for i in pic_urls:
                pic_info = i.get_attribute('style')
                pic_url = re.search(r'(https://.+)_tn',pic_info).group(1)
                pic_name = 'pro_{}.jpg'.format(n)
                n += 1
                save_image(pic_url,item_name,pic_name,save_path)
            

            variations = driver.find_elements_by_css_selector('.product-variation')
            disabled = driver.find_elements_by_css_selector('.product-variation.product-variation--disabled')
            variations_list = set([i for i in variations])
            disabled_list = set([i for i in disabled])
            enabled_list = variations_list-disabled_list
            variation_info = ''
            if len(enabled_list) != 0:
                for i in enabled_list:
                    print('getting variations and its prices')
                    i.click()
                    vprice = driver.find_element_by_css_selector('div._3n5NQx').text
                    variation = i.text
                    print('getting current price of {} : {}'.format(variation,vprice))

                    try:
                        vlist_price = driver.find_element_by_css_selector('div._3_ISdg').text
                        print('getting list price: {}'.format(vlist_price))
                    except NoSuchElementException:
                        vlist_price = vprice
                        print('list price is equal to price')
                    variation_info += variation+':'+vprice+ ':'+vlist_price+':'
            else:
                print('this product has no variations')
                variation = 'N/A'
                
                vprice = driver.find_element_by_css_selector('div._3n5NQx').text
                print('getting current price {}'.format(vprice))
                try:
                    vlist_price = driver.find_element_by_css_selector('div._3_ISdg').text
                    print('getting list price: {}'.format(vlist_price))
                except NoSuchElementException:
                    vlist_price = vprice
                    print('list price is equal to price')
                variation_info += variation+':'+vprice+ ':'+vlist_price


            try:
                driver.save_screenshot(r'shop_crawl\shopee\fullscreen.png')
                description = driver.find_element_by_css_selector('div.product-detail.page-product__detail')
                left = description.location['x']
                top = description.location['y']
                right = description.location['x'] + description.size['width']
                bottom = description.location['y'] + description.size['height']
                im = Image.open(r'shop_crawl\shopee\fullscreen.png')
                im = im.crop((left, top, right, bottom))
                im.save(r'shop_crawl\shopee\{}\des.png'.format(item_name))
     
            except NoSuchElementException as err:
                print('did not get description Error message:{}'.format(err))

            stock = 1000
            warning_code = 10
            shipping_fee = 15
            w.write('{},{},{},{},{},{},{},{},{},{},{},{}\n'.format(platform,category_info,item_name,brand,list_price,price,variation_info,after_sale,stock,warning_code,shipping_fee,item_url))
            driver.close()
            
        except TimeoutException as err:
            print(err)
            print('failed to load this url: {}, waitting for 10 seconds to reload'.format(item_url))
            time.sleep(10)
            get_shopee_details(item_url,save_file)
            
            
    print('getting item details done')

    
def save_image(img_url,director,pic_name,save_path):
    print('{} saving picture... {}'.format(pic_name,img_url))
    img_response = requests.get(img_url,headers=headers)
    path = save_path.format(director)
    isExists = os.path.exists(path)
    if not isExists:
        os.makedirs(path)
    with open(r'{}\{}'.format(path,pic_name),'ab') as f:
        f.write(img_response.content)
        

def shopee(file,save_file,save_path):
    start = time.time()
    
    with open(save_file,'a',encoding='utf-8,', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Platform','Category','Item Name','Brand','List Price','Price','Specification','Warranty Service','Stock','Warning Code','Shipping Fee','Item Url'])
    
    with open(file,'r',encoding='utf-8') as f:
        i = 1
        reader = csv.reader(f)
        rows = [row for row in reader]
        for row in rows:
            print('Getting item No.{}'.format(i))
            item_url = row[0]
            get_shopee_details(item_url,save_file)
            i += 1
            time.sleep(10)
    end = time.time()
    print('End in {}'.format(end-start))
        
        
#item_url = "https://shopee.co.id/Jam-Tangan-Pria-SKMEI-S-Shock-Dual-Time-0993-i.3118593.68780735"
save_file = r'shop_crawl\shopee\shopee_nero.csv'
save_path = r'shop_crawl\shopee\{}'
file = r'shop_crawl\nero.csv'

shopee(file,save_file,save_path)
```

```python
from selenium import webdriver
import time,re,csv,random,datetime,requests,os
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains
from PIL import Image
from selenium.webdriver.common.keys import Keys
from pyquery import PyQuery as pq

headers = {'User-Agent': 'User-Agent:Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}

def get_lazada_details(item_url,save_file):

    with open(save_file,'a',encoding='utf-8') as w:
        try:
            print('getting item details from Lazada ...')
            driver = webdriver.PhantomJS()
            driver.maximize_window()
            driver.get(item_url)
            time.sleep(10) 
            print('item url: {}'.format(item_url))
            html = driver.page_source
            doc = pq(html)
            category = doc('#J_breadcrumb')
            item_class = category.find('.breadcrumb_item')
            category_info = re.sub(',',' ',':'.join([i.text() for i in item_class.items()]),count=0)  
            print('getting category list: {}'.format(category_info))
            item = doc('.pdp-block.pdp-block__main-information-detail')
            item_title = item.find('.pdp-product-title').text()
            item_name = ' '.join(re.sub('[^a-z0-9A-Z_\-\s]','', item_title, count=0).split()) 
            print('getting item title:{}'.format(item_name)) 
            brand = item.find('.pdp-link.pdp-link_size_s.pdp-link_theme_blue.pdp-product-brand__brand-link').text()
            print('getting brand: {}'.format(brand)) 

            price_info = item.find('.pdp-product-price')
            price_info.find('.pdp-product-price__discount').remove()
            price = re.sub('\n',' ',price_info.text()).split(' ')
            current_price = price[0]
            print('getting current price: {}'.format(current_price))
            list_price = price[1]
            print('getting list price:{}'.format(list_price))            
            

            warranty = re.sub('\n','',item.find('.warranty').text(),count=0)
            print('getting warranty service: {}'.format(warranty))
            try:
                sku = item.find('.sku-prop')
                variations = ':'.join([i.find('.section-title').text() for i in sku.items()])
                print('getting variations:{}'.format(variations))
            except NoSuchElementException:
                print('this product has no variations')
                variations = 'N/A'


            # Product pictures

            pic_urls = driver.find_elements_by_css_selector('.item-gallery__image-wrapper img')
            n = 1
            for i in pic_urls:
                pic_url = i.get_attribute('src')
                pic_url = re.sub('80x80','800x800',pic_url,count=0)
                pic_url = re.sub('_.webp','',pic_url,count=0)
                pic_name = 'pro_{}.jpg'.format(n)
                n += 1
                save_image(pic_url,item_name,pic_name,save_path)


            # specification pictures
            try:
                specification_pic = driver.find_elements_by_css_selector('.sku-prop-content')
                for s in specification_pic:
                    imgs = s.find_elements_by_tag_name('img')
                    b = 1
                    for img in imgs:
                        spe_url = img.get_attribute('src')
                        spe_url = re.sub('80x80','800x800',spe_url,count=0)
                        spe_url = re.sub('_.webp','',spe_url,count=0)
                        spe_name = 'spe_{}.jpg'.format(b)
                        b += 1
                        save_image(spe_url,item_name,spe_name,save_path)
            except:
                print('This product has no pictures in specification.')                
            
            
            # description pictures
            
            try:
                print('printing screenshot...')
                driver.save_screenshot(r'shop_crawl\lazada\fullscreen.png')
                description = driver.find_element_by_id('module_product_detail')
                left = description.location['x']
                top = description.location['y']
                right = description.location['x'] + description.size['width']
                bottom = description.location['y'] + description.size['height']
                im = Image.open(r'shop_crawl\lazada\fullscreen.png')
                im = im.crop((left, top, right, bottom))
                im.save(r'shop_crawl\lazada\{}\des.png'.format(item_name))
            except NoSuchElementException as err:
                print('did not get description Error message:{}'.format(err))            
            
            platform = 'Lazada'
            stock = 1000
            warning_code = 10
            shipping_fee = 15
            w.write('{},{},{},{},{},{},{},{},{},{},{},{}\n'.format(platform,category_info,item_name,brand,list_price,current_price,variations,warranty,stock,warning_code,shipping_fee,item_url))       
            driver.close()
            
        except TimeoutException as err:
            print(err)
            print('failed to load this url: {}, waitting for 10 seconds to reload'.format(item_url))
            time.sleep(10)
            get_lazada_details(item_url,save_file)
    print('getting item details done')
    
def save_image(img_url,director,name,save_path):
    print('saving picture  {} ...'.format(name))
    img_response = requests.get(img_url,headers=headers)
    path = save_path.format(director)
    isExists = os.path.exists(path)
    if not isExists:
        os.makedirs(path)
    with open('{}\{}'.format(path,name),'ab') as f:
        f.write(img_response.content)


def lazada(file,save_file,save_path):
    start = time.time()
    
    with open(save_file,'a',encoding='utf-8,', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Platform','Category','Item Name','Brand','List Price','Price','Specification','Warranty Service','Stock','Warning Code','Shipping Fee','Item Url'])

    
    with open(file,'r',encoding='utf-8') as f:
        i = 1
        reader = csv.reader(f)
        rows = [row for row in reader]
        for row in rows[10:]:
            print('Getting item No.{}'.format(i))
            item_url = row[0]
            get_lazada_details(item_url,save_file)
            i += 1
            time.sleep(10)
    end = time.time()
    print('End in {}'.format(end-start))        

        
#item_url = 'https://www.lazada.co.id/products/jaket-parka-taslan-waterproof-oenzie-store-i196852035-s394004053.html?spm=a2o4j.searchlistcategory.list.1.55293f0a2OxWHU&search=1'
save_file = r'shop_crawl\lazada\lazada_result.csv'
save_path = r'shop_crawl\lazada\{}'
file = r'shop_crawl\lazada\lazada_test.csv'

lazada(file,save_file,save_path)
```

```python
from selenium import webdriver
import time,re,csv,random,datetime,requests,os, pymongo
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import ActionChains
from PIL import Image
from selenium.webdriver.common.keys import Keys
from pyquery import PyQuery as pq

headers = {'User-Agent': 'User-Agent:Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}

def get_shopee_details(item_url,save_path):
    
    try:
        print('getting item details from Shopee ...')
        driver = webdriver.PhantomJS()
        driver.maximize_window()
        driver.get(item_url)
        wait = WebDriverWait(driver,10)
        wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.flex.items-center._1z1CEl.page-product__breadcrumb')))
        item_class = driver.find_elements_by_css_selector('div.flex.items-center._1z1CEl.page-product__breadcrumb a')
        print('item url: {}'.format(item_url))
        platform = item_class[0].text
        print('getting item platform: {}'.format(platform))
        main_class = item_class[1].text
        print('getting item main class: {}'.format(main_class))
        sub_class = item_class[2].text
        print('getting item sub class: {}'.format(sub_class))
        category_info = [i.text for i in item_class]
        print('getting item category：{}'.format(category_info))
        pro_name = driver.find_element_by_css_selector('div.flex.items-center._1z1CEl.page-product__breadcrumb span').text
        pro_name = re.sub('shopee','',pro_name,count=0,flags=re.I)
        item_name = ' '.join(re.sub('[^a-z0-9A-Z_\-\s]','', pro_name, count=0).split())
        print('getting pro name: {}'.format(pro_name))

        for i in range(5):
            time.sleep(1)
            driver.find_element_by_xpath("/html/body").send_keys(Keys.PAGE_DOWN)            

        try:
            aftersale_info = driver.find_element_by_css_selector('a._13C8_x.flex.items-center').text
            after_sale = re.sub(r'\n',' ',aftersale_info,count=0)
            print('getting after sale info: {}'.format(after_sale))
            
        except:
            try:
                aftersale_info = driver.find_element_by_css_selector('div._3E_omo').text
                after_sale = re.sub(r'\n',' ',aftersale_info,count=0)
                print('getting after sale info: {}'.format(after_sale))
            except Exception as err:
                after_sale = 'N/A'
                print('did not get after sale info Error message:{}'.format(err))

        try:
            brand = driver.find_element_by_css_selector('a._2H-513').text
            print('getting brand: {}'.format(brand))
        except Exception as err:
            brand = 'N/A'
            print('did not get brand Error message:{}'.format(err))

        price = driver.find_element_by_css_selector('div._3n5NQx').text
        print('getting current price: {}'.format(price))

        try:
            list_price = driver.find_element_by_css_selector('div._3_ISdg').text
            print('getting list price: {}'.format(list_price))
        except NoSuchElementException:
            list_price = price
            print('list price is equal to price')


        print('getting pic urls')
        pic_urls = driver.find_elements_by_css_selector('._2Fw7Qu.animated-lazy-image__image--ready')
        n = 1
        pro_pictures = []
        
        for i in pic_urls:
            pic_info = i.get_attribute('style')
            pic_url = re.search(r'(https://.+)_tn',pic_info).group(1)
            pic_name = 'pro_{}.jpg'.format(n)
            pro_pictures.append(pic_url)
            n += 1
            save_image(pic_url,item_name,pic_name,save_path)
        
        
        variations = driver.find_elements_by_css_selector('.product-variation')
        disabled = driver.find_elements_by_css_selector('.product-variation.product-variation--disabled')
        variations_list = set([i for i in variations])
        disabled_list = set([i for i in disabled])
        enabled_list = variations_list-disabled_list
        variation_info = []
        if len(enabled_list) != 0:
            for i in enabled_list:
                print('getting variations and its prices')
                i.click()
                vprice = driver.find_element_by_css_selector('div._3n5NQx').text
                variation = i.text
                print('getting current price of {} : {}'.format(variation,vprice))

                try:
                    vlist_price = driver.find_element_by_css_selector('div._3_ISdg').text
                    print('getting list price: {}'.format(vlist_price))
                except NoSuchElementException:
                    vlist_price = vprice
                    print('list price is equal to price')
                    
                variation_dic = {
                    'variation_name': variation,
                    'variation_price': vprice,
                    'variation_listprice': vlist_price
                }
                variation_info.append(variation_dic)
                
        else:
            print('this product has no variations')
            variation = 'N/A'

            vprice = driver.find_element_by_css_selector('div._3n5NQx').text
            print('getting current price {}'.format(vprice))
            try:
                vlist_price = driver.find_element_by_css_selector('div._3_ISdg').text
                print('getting list price: {}'.format(vlist_price))
            except NoSuchElementException:
                vlist_price = vprice
                print('list price is equal to price')
            variation_dic = {
                'variation_name': variation,
                'variation_price': vprice,
                'variation_listprice': vlist_price
            }
            variation_info.append(variation_dic)            

        try:
            print('printing screenshot...')
            driver.save_screenshot(fullscreen)
            description = driver.find_elements_by_css_selector('div._2C2YFD')[-1]
            left = description.location['x']
            top = description.location['y']
            right = description.location['x'] + description.size['width']
            bottom = description.location['y'] + description.size['height']
            im = Image.open(fullscreen)
            h = description.size['height']
            if h <= height:
                im = im.crop((left, top, right, bottom))
                im.save((save_path + '\des.png').format(item_name))
                print('getting description pic des.png')
            else:
                times = h//height 
                mod = h%height
                top = description.location['y']
                bottom = description.location['y'] + height
                for i in range(times):
                    piece = im.crop((left, top, right, bottom))
                    path = save_path + '\des_{}.png'.format(i+1)
                    piece.save(path.format(item_name))                    
                    top = bottom
                    bottom = bottom + height
                    print('getting description pic des_{}.png'.format(i+1))
                piece = im.crop((left, top, right, top+mod))
                path = save_path + '\des_{}.png'.format(times+1)
                piece.save(path.format(item_name)) 
                print('getting description pic des_{}.png'.format(times+1))
        except NoSuchElementException as err:
            print('did not get description Error message:{}'.format(err))

        product = {
            'platform': platform,
            'category_info': category_info,
            'item_name': pro_name,
            'brand': brand,
            'list_price': list_price,
            'price': price,
            'variation_info': variation_info,
            'warranty': after_sale,
            'pic_urls': pro_pictures,
            'item_url': item_url,
            'stock': 1000,
            'warning_code': 10,
            'shipping_fee': 15
        }

        save_to_mongo(product)

        driver.close()
            
    except TimeoutException as err:
        print(err)
        print('failed to load this url: {}, waitting for 10 seconds to reload'.format(item_url))
        time.sleep(10)
        get_shopee_details(item_url,save_path)

    print('getting item details done')

    
def get_lazada_details(item_url,save_path):

    try:
        print('getting item details from Lazada ...')
        driver = webdriver.PhantomJS()
        driver.maximize_window()
        driver.get(item_url)
        time.sleep(10) 
        print('item url: {}'.format(item_url))
        html = driver.page_source
        doc = pq(html)
        category = doc('#J_breadcrumb')
        item_class = category.find('.breadcrumb_item')
        category_info = [i.text() for i in item_class.items()]
        print('getting category list: {}'.format(category_info))
        item = doc('.pdp-block.pdp-block__main-information-detail')
        item_title = item.find('.pdp-product-title').text()
        item_name = ' '.join(re.sub('[^a-z0-9A-Z_\-\s]','', item_title, count=0).split()) 
        print('getting item title:{}'.format(item_name)) 
        brand = item.find('.pdp-link.pdp-link_size_s.pdp-link_theme_blue.pdp-product-brand__brand-link').text()
        print('getting brand: {}'.format(brand)) 

        price_info = item.find('.pdp-product-price')
        price_info.find('.pdp-product-price__discount').remove()
        price = re.sub('\n',' ',price_info.text()).split(' ')
        if len(price) == 2:
            current_price = price[0]
            print('getting current price: {}'.format(current_price))
            list_price = price[1]
            print('getting list price:{}'.format(list_price))            
        elif len(price) == 1:
            current_price = price[0]
            print('getting current price: {}'.format(current_price))
            list_price = current_price
            print('list price is equal to price')                 

        warranty = re.sub('\n','',item.find('.warranty').text(),count=0)
        print('getting warranty service: {}'.format(warranty))
        try:
            sku = item.find('.sku-prop')
            variations = [i.find('.section-title').text() for i in sku.items()]
            print('getting variations:{}'.format(variations))
        except NoSuchElementException:
            print('this product has no variations')
            variations = 'N/A'


        # Product pictures
        pro_pictures = []
        pic_urls = driver.find_elements_by_css_selector('.item-gallery__image-wrapper img')
        n = 1
        for i in pic_urls:
            pic_url = i.get_attribute('src')
            pic_url = re.sub('80x80','800x800',pic_url,count=0)
            pic_url = re.sub('_.webp','',pic_url,count=0)
            pic_name = 'pro_{}.jpg'.format(n)
            pro_pictures.append(pic_url)
            n += 1
            save_image(pic_url,item_name,pic_name,save_path)


        # specification pictures
        spe_pictures = []
        try:
            specification_pic = driver.find_elements_by_css_selector('.sku-prop-content')
            for s in specification_pic:
                imgs = s.find_elements_by_tag_name('img')
                b = 1
                for img in imgs:
                    spe_url = img.get_attribute('src')
                    spe_url = re.sub('80x80','800x800',spe_url,count=0)
                    spe_url = re.sub('_.webp','',spe_url,count=0)
                    spe_pictures.append(spe_url)
                    spe_name = 'spe_{}.jpg'.format(b)
                    b += 1
                    save_image(spe_url,item_name,spe_name,save_path)
        except:
            print('This product has no pictures in specification.')                


        # description pictures

        try:
            print('printing screenshot...')
            driver.save_screenshot(fullscreen)
            description = driver.find_element_by_id('module_product_detail')
            left = description.location['x']
            top = description.location['y']
            right = description.location['x'] + description.size['width']
            bottom = description.location['y'] + description.size['height']
            im = Image.open(fullscreen)
            h = description.size['height']
            if h <= height:
                im = im.crop((left, top, right, bottom))
                im.save((save_path + '\des.png').format(item_name))
                print('getting description pic des.png')
            else:
                times = h//height 
                mod = h%height
                top = description.location['y']
                bottom = description.location['y'] + height
                for i in range(times):
                    piece = im.crop((left, top, right, bottom))
                    path = save_path + '\des_{}.png'.format(i+1)
                    piece.save(path.format(item_name))                    
                    top = bottom
                    bottom = bottom + height
                    print('getting description pic des_{}.png'.format(i+1))
                piece = im.crop((left, top, right, top+mod))
                path = save_path + '\des_{}.png'.format(times+1)
                piece.save(path.format(item_name)) 
                print('getting description pic des_{}.png'.format(times+1))
        except NoSuchElementException as err:
            print('did not get description Error message:{}'.format(err))            
 
            
        platform = 'Lazada'
        product = {
            'platform': platform,
            'category_info': category_info,
            'item_name': item_title,
            'brand': brand,
            'list_price': list_price,
            'price': current_price,
            'variation_info': variations,
            'warranty': warranty,
            'item_url': item_url,
            'pic_urls': pro_pictures,
            'spe_urls': spe_pictures,                
            'stock': 1000,
            'warning_code': 10,
            'shipping_fee': 15
        }            
        save_to_mongo(product)            
        driver.close()

    except TimeoutException as err:
        print(err)
        print('failed to load this url: {}, waitting for 10 seconds to reload'.format(item_url))
        time.sleep(10)
        get_lazada_details(item_url,save_path)
        
    print('getting item details done')
    
    
def get_myntra_details(item_url,save_path):

    try:
        print('getting item details from myntra ...')
        driver = webdriver.PhantomJS()
        driver.maximize_window()
        driver.get(item_url)
        time.sleep(2) 
        print('item url: {}'.format(item_url))
        html = driver.page_source
        doc = pq(html)

        item = doc('.pdp-pdp-container')
        category = item.find('.breadcrumbs-container a')
        category_info = [i.text() for i in category.items()][:-1]
        print('getting category list: {}'.format(category_info))

        brand = item.find('.pdp-title').text()
        print('getting brand: {}'.format(brand))

        item_title = brand + ' ' + item.find('.pdp-name').text() 
        item_name = ' '.join(re.sub('[^a-z0-9A-Z_\-\s]','', item_title, count=0).split())
        print('getting item title:{}'.format(item_name)) 

        price = item.find('.pdp-price').text()
        print('getting current price: {}'.format(price))
        list_price = item.find('.pdp-mrp').text()
        if list_price != '':
            print('getting list price:{}'.format(list_price))
        else:
            list_price = price
            print('list price is equal to price')

        sku = item.find('.size-buttons-unified-size')  
        variations = [i.text() for i in sku.items()]
        print('getting variations:{}'.format(variations))            

        try:
            driver.find_element_by_css_selector('.index-showMoreText').click()
            print('click button to see more')
        except NoSuchElementException:
            print('there is no see more button')
            pass

        # Product pictures
        pro_pictures = []
        pic_urls = driver.find_elements_by_css_selector('.image-grid-image')
        n = 1
        for i in pic_urls:
            pic_url = re.search(r'url\((.+)\)',i.get_attribute('style')).group(1)
            pro_pictures.append(pic_url)
            pic_name = 'pro_{}.jpg'.format(n)
            n += 1
            save_image(pic_url,item_name,pic_name,save_path)            

        # description pictures
        time.sleep(10) 
        
        
        try:
            print('printing screenshot...')
            driver.save_screenshot(fullscreen)
            description = driver.find_element_by_css_selector('.pdp-productDescriptorsContainer')
            left = description.location['x']
            top = description.location['y']
            right = description.location['x'] + description.size['width']
            bottom = description.location['y'] + description.size['height']
            im = Image.open(fullscreen)
            h = description.size['height']
            if h <= height:
                im = im.crop((left, top, right, bottom))
                im.save((save_path + '\des.png').format(item_name))
                print('getting description pic des.png')
            else:
                times = h//height 
                mod = h%height
                top = description.location['y']
                bottom = description.location['y'] + height
                for i in range(times):
                    piece = im.crop((left, top, right, bottom))
                    path = save_path + '\des_{}.png'.format(i+1)
                    piece.save(path.format(item_name))                    
                    top = bottom
                    bottom = bottom + height
                    print('getting description pic des_{}.png'.format(i+1))
                piece = im.crop((left, top, right, top+mod))
                path = save_path + '\des_{}.png'.format(times+1)
                piece.save(path.format(item_name)) 
                print('getting description pic des_{}.png'.format(times+1))
        except NoSuchElementException as err:
            print('did not get description Error message:{}'.format(err))          
      

        platform = 'Myntra'
        warranty = '100% ORIGINAL guarantee for all products at myntra.com;Return within 30days of receiving your order;Get free delivery for every order above Rs.1199'

        product = {
            'platform': platform,
            'category_info': category_info,
            'item_name': item_title,
            'brand': brand,
            'list_price': list_price,
            'price': price,
            'variation_info': variations,
            'warranty': warranty,
            'item_url': item_url,
            'pic_urls': pro_pictures,
            'stock': 1000,
            'warning_code': 10,
            'shipping_fee': 15
        }
        save_to_mongo(product)            

        driver.close()

    except TimeoutException as err:
        print(err)
        print('failed to load this url: {}, waitting for 10 seconds to reload'.format(item_url))
        time.sleep(10)
        get_myntra_details(item_url,save_path)
        
    print('getting item details done')    
    
    
def save_image(img_url,director,pic_name,save_path):
    print('{} saving picture... {}'.format(pic_name,img_url))
    img_response = requests.get(img_url,headers=headers)
    path = save_path.format(director)
    isExists = os.path.exists(path)
    if not isExists:
        os.makedirs(path)
    with open(r'{}\{}'.format(path,pic_name),'ab') as f:
        f.write(img_response.content)

        
def save_to_mongo(result):
    MONGO_URL = 'localhost'
    MONGO_DB = 'shopcrawl'
    MONGO_COLLECTION = 'products'
    client = pymongo.MongoClient(MONGO_URL)
    db = client[MONGO_DB]
    collection = db[MONGO_COLLECTION]
    
    try:
        if collection.insert_one(result):
            print('save to mongodb')
    except Exception:
        print('failed to save to mongodb')
        
        
        
def shop_crawl (file,save_path):
    start = time.time()

    with open(file,'r',encoding='utf-8') as f:
        i = 1
        reader = csv.reader(f)
        rows = [row for row in reader]
        for row in rows:
            item_url = row[0]
            print('Getting item No.{}'.format(i))
            if 'shopee' in item_url:
                get_shopee_details(item_url,save_path)
                i += 1
                time.sleep(10)
            elif 'lazada' in item_url:
                get_lazada_details(item_url,save_path)
                i += 1
                time.sleep(10)                
            elif 'myntra' in item_url:
                get_myntra_details(item_url,save_path)
                i += 1
                time.sleep(10)        
    end = time.time()
    print('End in {} minutes'.format((end-start)/60))



stock = 1000
warning_code = 10
shipping_fee = 15
height = 1000

save_path = r'shop_crawl\shopcrawl\test0114\{}'
file = r'shop_crawl\shopcrawl\test0114.csv'
fullscreen = r'shop_crawl\shopcrawl\test0114\fullscreen.png'

shop_crawl(file,save_path)
```
