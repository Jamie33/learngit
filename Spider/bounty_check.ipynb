```python
from pyquery import PyQuery as pq
import csv


filename = r'NAUS ALL IN SPREADSHEAT - TWITTER0814.csv'
with open('Twiiter Audit Stake0814.csv','a',encoding='utf-8') as w:
    with open(filename,'r',encoding='utf-8') as f:
        reader = csv.reader(f)
        urls = [row[5] for row in reader]
        for url in urls[1:]:
            try:
                doc = pq(url)
                number = int(doc('.color--green-base .real.number').attr('data-value'))
                if number > 10000:
                    stake = 8
                elif number > 5000:
                    stake = 5
                elif number > 3000:
                    stake = 3
                elif number > 1000:
                    stake = 2
                elif number>=20 or number<=1000: 
                    stake = 1    
            except:
                number ='wrong audit url'
                stake = 0
            print(url,number,stake)
            w.write('{},{},{}\n'.format(url,number,stake))
print('done')
```
