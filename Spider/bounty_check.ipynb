```python
from pyquery import PyQuery as pq
import csv

def check_number(url):
    try:
        doc = pq(url)
        number = int(doc('.color--green-base .real.number').attr('data-value'))
        status = 'Accepted'
        if number > 10000:
            stake = 8
        elif number > 5000:
            stake = 5
        elif number > 3000:
            stake = 3
        elif number > 1000:
            stake = 2
        elif number>=20 and number<=1000: 
            stake = 1    
    except:
        number ='wrong audit url'
        stake = 0
        status = 'Rejected'

        


filename = r'NAUS ALL IN SPREADSHEAT - TWITTER.csv'
with open('Twiiter Audit Stake.csv','a',encoding='utf-8') as w:
    with open(filename,'r',encoding='utf-8') as f:
        reader = csv.reader(f)
        urls = [row[5] for row in reader]
        for url in urls[1:]:
            try:
                doc = pq(url)
                number = int(doc('.color--green-base .real.number').attr('data-value'))
                if number > 10000:
                    stake = 8
                elif number > 5000:
                    stake = 5
                elif number > 3000:
                    stake = 3
                elif number > 1000:
                    stake = 2
                elif number>=20 or number<=1000: 
                    stake = 1    
            except:
                number ='wrong audit url'
                stake = 0
            w.write('{},{},{}\n'.format(url,number,status,stake))
print('done')
```

```python
import csv 


filename = r'NAUS ALL IN SPREADSHEAT - TWITTER.csv'

with open('twitterfollowerverfiy.csv','a',encoding='utf-8') as w:
    with open('twitterfollowers.txt','r',encoding='utf-8') as r:
        content = r.read()
        with open(filename,'r',encoding='utf-8') as f:
            reader = csv.reader(f)
            urls = [row[5] for row in reader]
            for url in urls[1:]:
                name = url.split('/')[-1].strip('')
                if name in content:
                    status = 'follow'
                else:
                    status = 'not follow'
                w.write('{},{},{}\n'.format(url,name,status))
print('done')
```

```python
from selenium import webdriver
import time
import re
from requests_html import HTMLSession
session = HTMLSession()
proxie = {'https':'https://127.0.0.1:1080'}

url = 'https://www.facebook.com/Daniel.Hunter686'

options = webdriver.ChromeOptions()
options.add_argument('--proxy-server=http://127.0.0.1:1080')  
driver = webdriver.Chrome(chrome_options=options)

driver.get(url)

time.sleep(1)

target = driver.find_element_by_xpath('//*[@id="u_fetchstream_1_0"]/div[1]/div/div/div[2]/div/span[2]/a')

for i in range(10):
    driver.execute_script('arguments[0].scrollIntoView();',target)
    time.sleep(1)

driver.implicitly_wait(10)

tweet = driver.find_elements_by_css_selector('.js-tweet-text-container')
time = driver.find_elements_by_css_selector('._timestamp')
fullname = driver.find_elements_by_css_selector('.FullNameGroup')

driver.implicitly_wait(10)


address = r'(0x[a-z0-9A-Z]{40})'


with open('twitter20180503.csv','w',encoding='utf-8') as f:
    for i in range(len(tweet)):
        if re.search(address,tweet[i].text):
            eth = re.search(address,tweet[i].text).group(0)
            date = time[i].text
            name = fullname[i].text
            f.write('{},{},{}\n'.format(name.strip(),date.strip(),eth.strip()))

print('The END')              
    
driver.quit()
```
