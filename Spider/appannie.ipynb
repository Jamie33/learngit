# App annie 应用商店排行榜
日本，电商类,ios 和 google play

```python
from selenium import webdriver
import time
from pyquery import PyQuery as pq

def appann(username,password,rank_url,file_name):
    with open(file_name,'a',encoding='utf-8') as w:
        url = 'https://www.appannie.com/account/login/?_ref=header'
        driver = webdriver.Chrome()  #声明浏览器对象
        driver.get(url) # 访问网页
        driver.find_element_by_id('email').send_keys(username) # 提取账号输入框，并输入用户名
        driver.find_element_by_id('password').send_keys(password) # 提取密码输入框，并输入密码
        driver.find_element_by_id('submit').submit() # 点击登陆按钮 
        driver.get(rank_url) # 访问网页
        time.sleep(10)
        content =driver.find_elements_by_css_selector('tr.main-row.table-row')
        for i in content:
            name = i.find_element_by_css_selector('.app-link').text.replace(',','')
            company_name = i.find_element_by_css_selector('.company-link').text.replace(',','')
            print(name,company_name)
            w.write('{},{}\n'.format(name,company_name))
        print('done')
        driver.quit() # 退出浏览器

username = '' # 你的账号
password = '' # 你的密码
ios_url = 'https://www.appannie.com/apps/ios/top-chart/?country=JP&category=6024&device=iphone&date=2018-11-30&feed=All&rank_sorting_type=rank&page_number=0&page_size=100&table_selections=&order_type=desc&order_by=free_rank'
ios_filename = 'annie_ios_shopping.csv'
google_url  ='https://www.appannie.com/apps/google-play/top-chart/?country=JP&category=30&device=&date=2018-11-30&feed=All&rank_sorting_type=rank&page_number=0&page_size=100&table_selections='
google_filename = 'annie_googleplay_shopping.csv'

appann(username,password,ios_url,ios_filename)
#appann(username,password,google_url,google_filename)
```

```python
from selenium import webdriver
import time
from pyquery import PyQuery as pq

def appann(username,password,rank_url,file_name,keyword_str):
    with open(file_name,'a',encoding='utf-8') as w:
        url = 'https://www.appannie.com/account/login/?_ref=header'
        driver = webdriver.Chrome()  #声明浏览器对象
        driver.get(url) # 访问网页
        driver.find_element_by_id('email').send_keys(username) # 提取账号输入框，并输入用户名
        driver.find_element_by_id('password').send_keys(password) # 提取密码输入框，并输入密码
        driver.find_element_by_id('submit').submit() # 点击登陆按钮 
        driver.get(rank_url) # 访问网页
        time.sleep(10)
        content =driver.find_elements_by_css_selector('tr.main-row.table-row')
        for i in content:
            name = i.find_element_by_css_selector('.app-link').text.replace(',','')
            app_url = i.find_element_by_css_selector('.app-link').get_attribute('href')
            kw_url = app_url[0:-8] + keyword_str
            print(name,app_url,kw_url)
            w.write('{},{},{}\n'.format(name,app_url,kw_url))
        print('done')
        driver.quit() # 退出浏览器

username = '' # 你的账号
password = '' # 你的密码
ios_url = 'https://www.appannie.com/apps/ios/top-chart/?country=JP&category=6024&device=iphone&date=2018-11-30&feed=All&rank_sorting_type=rank&page_number=0&page_size=100&table_selections=&order_type=desc&order_by=free_rank'
ios_filename = 'annie_ios_shopping_appurl.csv'
ios_keyword_str = 'keywords/?device=iphone&date=2018-09-04~2018-12-03&countries=JP'
google_url  ='https://www.appannie.com/apps/google-play/top-chart/?country=JP&category=30&device=&date=2018-11-30&feed=All&rank_sorting_type=rank&page_number=0&page_size=100&table_selections='
google_filename = 'annie_googleplay_shopping_appurl.csv'
google_keyword_str = 'keywords/?device=android&date=2018-09-06~2018-12-04&countries=JP&order_by=isPinned&order_type=desc&chart_type=rank&table_selections=1797759&series=1797759'

appann(username,password,ios_url,ios_filename,ios_keyword_url)
#appann(username,password,google_url,google_filename,google_keyword_str)
```

```python
from selenium import webdriver
import time,csv,random
from pyquery import PyQuery as pq

def appann(username,password,file_name,save_file):
    url = 'https://www.appannie.com/account/login/?_ref=header'
    driver = webdriver.Chrome()  #声明浏览器对象
    driver.get(url) # 访问网页
    driver.find_element_by_id('email').send_keys(username) # 提取账号输入框，并输入用户名
    driver.find_element_by_id('password').send_keys(password) # 提取密码输入框，并输入密码
    driver.find_element_by_id('submit').submit() # 点击登陆按钮 
    time.sleep(10)
    count = 1
    with open(save_file,'a',encoding='utf-8') as w:
        with open(file_name,'r',encoding='utf-8') as f:
            
            print(count)
            
            reader = csv.reader(f)
            rows = [row for row in reader]
            for row in rows[70:]:
                count += 1
                app_name = row[0]
                kw_url = row[2]
                driver.get(kw_url) # 访问网页
                time.sleep(random.choice(range(20,40)))
                keywords = driver.find_elements_by_css_selector('td.link2.tbl-col-link2')
                for kw in keywords:
                    print(app_name,kw.text)
                    w.write('{},{}\n'.format(app_name,kw.text))
   
    print('done')
    driver.quit() # 退出浏览器

username = '' # 你的账号
password = '' # 你的密码
ios_filename = 'annie_ios_shopping_appurl.csv' #打开文件
ios_save_file = 'annie_ios_shopping_keywords.csv' # 保存文件

google_filename = 'annie_googleplay_shopping_appurl.csv'
google_save_file = 'annie_google_shopping_keyword.csv'

appann(username,password,ios_filename,ios_save_file)
#appann(username,password,google_filename,google_save_file)
```
