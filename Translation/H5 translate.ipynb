# 提取文件中的中文字

```python
import re

def translate(filename,savefile):
    
    with open(savefile,'a',encoding='utf-8') as w:
        with open(filename,'r',encoding='utf-8') as f:
            htmlcont = f.read()
            rex = r'[\u4e00-\u9fa5]+'
            chlist = re.findall(rex,htmlcont)
            for ch in chlist:
                w.write('{}\n'.format(ch.replace(',','')))

filename = r'C:\Users\win7\learngit\Translation\views-cn\activity\index.html'
savefile = r'test1.csv'

translate(filename,savefile)
```

# 遍历文件下面的所有目录与文字

```python
import os 

file_list = []

def get_files(tar_dir):
    global file_list
    dir_list = os.listdir(tar_dir)  # #列出文件夹下所有的目录
    for i in dir_list:

        i_dir = os.path.join(tar_dir,i)
        if os.path.isdir(i_dir):
            get_files(i_dir)
        else:
            file_list.append(os.path.join(tar_dir,i_dir))
    
    return file_list


tar_dir = r'C:\Users\win7\learngit\Translation\views-cn'
get_files(tar_dir)

num = 0
for i in set(get_files(tar_dir)):
    num += 1
    print(i)
    print(num)
```
