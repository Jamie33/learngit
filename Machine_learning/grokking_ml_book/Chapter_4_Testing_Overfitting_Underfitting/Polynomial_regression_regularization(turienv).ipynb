# Chapter 4: Testing and regularization with Turi Create

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
random.seed(0)
```

## Defining our dataset

```python
# Our original polynomial is -x^2+x+15
coefs = [15,1,-1]
# y = -x^2+x+15 函数
def polynomial(coefs, x):   
    n = len(coefs)
    return sum([coefs[i]*x**i for i in range(n)])
# 绘制出该函数
def draw_polynomial(coefs):
    n = len(coefs)
    x = np.linspace(-5, 5, 1000)  # np.linspace() 生成指定范围内指定个数的一维数组
    plt.ylim(-20,20)
    plt.plot(x, polynomial(coefs, x), linestyle='-', color='black')

draw_polynomial(coefs)
```

```python
X = []
Y = []
for i in range(40):
    x = random.uniform(-5,5)  # 在 [x,y] 范围内随机生成一个实数
    y = polynomial(coefs, x) + random.gauss(0,2)  # 具有高斯分布的随机浮点数
    X.append(x)
    Y.append(y)
```

```python
plt.scatter(X, Y)
draw_polynomial(coefs)
```

## Training a polynomial regression model

```python
import turicreate as tc
```

```python
data = tc.SFrame({'x':X, 'y':Y})
data
```

### Adding columns to our dataset corresponding to $x^2, x^3, \ldots, x^{200}$

```python
for i in range(2,200):
    string = 'x^'+str(i)
    data[string] = data['x'].apply(lambda x:x**i)
data
```

## Splitting data into training and testing

**知识点  turicreate.SFrame.random_split(fraction, seed=None, exact=False)**
将 SFrame 的行随机拆分为两个 SFrame
- fraction 浮点数，取值范围 0-1。默认 exact = Fals，大概分割
- seed 随机种子
- exact 默认 False，如果为 True，会精准分割

```python
sf = tc.SFrame({'id': range(1024)})
sf_train, sf_test = sf.random_split(.9, seed=5) # 大概分割
print(len(sf_train), len(sf_test))
```

```python
sf = tc.SFrame({'id': range(1024)})
sf_train, sf_test = sf.random_split(.9,seed=5,exact=True) # 精准分割
print(len(sf_train), len(sf_test))
```

```python
train, test = data.random_split(.8, seed=0)
print(len(train), len(test))
```

```python
def display_results(model):
    coefs = model.coefficients # 获取模型的系数
    print("Training error (rmse):", model.evaluate(train)['rmse']) # 训练集 RMSE 均方根误差
    print("Testing error (rmse):", model.evaluate(test)['rmse'])  # 
    plt.scatter(train['x'], train['y'], marker='o')
    plt.scatter(test['x'], test['y'], marker='^')
    draw_polynomial(coefs['value'])
    plt.show()
    print("Polynomial coefficients")
    print(coefs['name', 'value'])
```

## Training a polynomial regression model with no regularization

```python
model_no_reg = tc.linear_regression.create(
    train, target='y', l1_penalty=0.0, l2_penalty=0.0, verbose=False, validation_set=None)
```

```python
display_results(model_no_reg)
```

## Training a polynomial regression model with L1 regularization

```python
model_L1_reg = tc.linear_regression.create(
    train, target='y', l1_penalty=0.1, l2_penalty=0.0, verbose=False, validation_set=None)
```

```python
display_results(model_L1_reg)
```

## Training a polynomial regression model with L2 regularization

```python
model_L2_reg = tc.linear_regression.create(
    train, target='y', l1_penalty=0.0, l2_penalty=0.1, verbose=False, validation_set=None)
```

```python
display_results(model_L2_reg)
```

## Viewing the predictions in the testing set

```python
predictions = test['x', 'y']
predictions['No reg'] = model_no_reg.predict(test)
predictions['L1 reg'] = model_L1_reg.predict(test)
predictions['L2 reg'] = model_L2_reg.predict(test)
```
