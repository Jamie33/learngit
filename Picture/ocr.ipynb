### 写入word 文档()

python-docx模块安装需要在cmd命令行中输入pip install python-docx

```python
import docx 
from docx.shared import Cm
import requests  # 库文件

def post_image(img):
    files = {"pic_path": open(img, "rb")}  # files # 类似data数据
    url = "http://pic.sogou.com/pic/upload_pic.jsp"  # post的url
    html = requests.post(url, files=files).text  # requests 提交图片
    return get_content(html)  # 结果是url就是图片的url sougou 把本地图片上传到sougou服务器变成了他的图片 调用解析函数把url传入
    
def get_content(keywords):
    url = "http://pic.sogou.com/pic/ocr/ocrOnline.jsp?query=" + keywords  # keywords就是图片url此方式为get请求
    ocrResult = requests.get(url).json()  # 直接转换为json格式
    contents = ocrResult['result']  # 类似字典 把result的value值取出来 是一个list然后里面很多json就是识别的文字
    text = ''
    for content in contents:  # 遍历所有结果
        text = text + content['content'].strip()  # strip去除空格 他返回的结果自带一个换行
    return text

def save_docx(text,img_name,file_name):
    print('识别文字',text)
    doc_new = docx.Document()  # 新建文档
    doc_new.add_picture(img_name,width=Cm(24))  # 向文档里添加图片
    doc_new.add_paragraph(text,style=None)  # 插入段落
    doc_new.save(file_name) # 保存文件
    print('图片和文字已保存到文件')

img = r"test2.jpg"  # 图片路径
file_name = r'ocr_file.docx'

text = post_image(img)
save_docx(text,img,file_name)
```
