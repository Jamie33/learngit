```python
import zipfile,os,re
f = zipfile.ZipFile(r'E:\桌面图片\channel.zip','r')
for file in f.namelist():
    f.extract(file,r'PC6zipfiles')
f.close()

comments = []

for i in os.listdir('PC6zipfiles'):
    if i == 'readme.txt':
        with open(os.path.join(os.getcwd(),'PC6zipfiles',i),'r') as txt:
            readme = txt.read()
            print(readme)
            hint = re.search(r'(\d){2,5}',readme).group()
            print(hint)
            comments.append(f.getinfo('%s.txt'%hint).comment)

while True:
    for a in os.listdir('PC6zipfiles'):
        if a == '%s.txt' % hint:
            with open(os.path.join(os.getcwd(),'PC6zipfiles',a),'r') as text:
                nothing = text.read()
                print(nothing,hint)
                if re.search(r'(\d){2,5}',nothing).group():
                    hint = re.search(r'(\d){2,5}',nothing).group()
                    comments.append(f.getinfo('%s.txt'%hint).comment)
```
